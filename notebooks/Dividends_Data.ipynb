### **Dividens**

```python
import numpy as np
import pandas as pd
import yfinance as yf
from datetime import datetime

tickers_AP = ['TLT', 'IEFA', 'PDBC', 'IAGG', 'IAU', 'DSMC', 'RSSB', 'ASFYX',
           'APDFX', 'QMHIX', 'CASH', 'GQGIX', 'SPY']
tickers_SPY_LV = ['SPY', 'CAOS']
tickers_SPY_LB = ['SPY', 'AGG']
tickers_GAL = ['GAL'] # State Street Global Allocation

tickers = [tickers_AP, tickers_SPY_LV, tickers_SPY_LB, tickers_GAL]
start_date = '2024-1-1'
end_date = datetime.now()
dividends_data = []

for ticker_group in tickers:
    group_divs = {}
    for ticker in ticker_group:
        hist = yf.Ticker(ticker).history(start = start_date, end = end_date)
        group_divs[ticker] = hist['Dividends']

    dividends_data.append(pd.DataFrame(group_divs))
```

```python
divs1 = dividends_data[0].copy()
divs2 = dividends_data[1].copy()
divs3 = dividends_data[2].copy()
divs4 = dividends_data[3].copy()
print(divs1.sample(5))
print('-'*50)
print(divs2.sample(5))
print('-'*50)
print(divs3.sample(5))
print('-'*50)
print(divs4.sample(5))
```
